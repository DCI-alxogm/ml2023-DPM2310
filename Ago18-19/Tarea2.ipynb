# Tarea 2

Encuentra un conjunto de datos que puedan ser descritos por un modelo lineal y usa el método de mínimos cuadrados para encontrar los parámetros de dicho modelo.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Consideramos los datos de ventas de un negocio en los ultimos 10 meses
data=np.loadtxt("negocio_alitas.txt",delimiter=',').T
```

```python
x=data[0]
y=data[1]
```

```python
plt.plot(x,y,'o')
plt.title('Venta de Alitas')
plt.legend()
plt.xlabel('Ventas $')
plt.ylabel('Mes')
plt.show()
```

```python
#Definimos los parametros necesarios para realizar el ajuste
N=len(x)
m=(N*sum(x*y)-sum(x)*sum(y))/(N*sum(x**2)-sum(x)**2)
b=(sum(y)-m*sum(x))/N

#Recta que mejor define los datos
f=m*x+b

#Graficamos
plt.plot(x,y,'o',label='(xi,yi)')
plt.plot(x,f,color='red',label='Ajuste Lineal')
plt.title('Minimos Cuadrados')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
